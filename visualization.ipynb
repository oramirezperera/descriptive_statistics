```
import pandas as pd
import seaborn as sns

iris = sns.load_dataset('iris')

iris.head()
```

sepal and petal length

```
# Scatter plot
sns.scatterplot(data = iris, x = 'sepal_length', y = 'petal_length', hue='species')
```

sepal and petal width

```
# scatter plot
sns.scatterplot(data = iris, x = 'sepal_width', y = 'petal_width', hue = 'species')
```

length

```
# joint plot
sns.jointplot(data = iris, x = 'sepal_length', y = 'petal_length', hue = 'species')
```

width

```
# jointplot
sns.jointplot(data = iris, x = 'sepal_width', y = 'petal_width', hue = 'species')
```

```
# Boxs plot
sns.boxplot(data = iris, x = 'species', y = 'sepal_length')
```

boxplot width

```
sns.boxplot(data = iris, x = 'species', y = 'sepal_width')
```

```
# bar plot
petal_length_mean = iris['petal_length'].mean()
petal_length_mean
sns.barplot(data = iris, x = 'species', y = 'petal_length')
```

```
sns.barplot(data = iris, x = 'species', y = 'petal_width')
```
